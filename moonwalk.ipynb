## Tacticus's Moon Hopper Mystery

"A moon-hopperâ€™s navigation boots are malfunctioning. Each thruster firing causes a fixed drift along a straight line:

Left thruster: 403 mm forward

Right thruster: 351 mm backward

The thrusters are perfectly accurate but cannot be adjusted. Each firing counts as one move, and you may fire either thruster any number of times, in any order.

You begin hovering exactly over a painted landing mark.

How close can you ever hover to the landing mark without being exactly on it?

What is the fewest number of thruster firings needed to achieve that closest approach?"

---


Well, both 403 and 351 are multiples of 13:

- 31 x 13 = 403
- 27 x 13 = 351

This means we can only ever move forward or backward by some multiple of 13 mm.

If we wanted, we could return to hover directly over the landing spot by taking 351 steps forward and 403 steps backward (Michael Jackson would have done this).

But the next closest hovering stations would be at some multiple of 13mm.

---

So let's simplify things and think in 13mm lengths.

A forward thrust moves 31 lengths and a backward thrust moves 27 lengths.

- f = 31
- b = -27

Can I find some combination of n and m such that nf + mb = 1 or -1?

- 31n - 27m = 1

let n = 7 and m = 8:

- 31(7) - 27(8) = 1

So, yes, by moving 7 thrusts forward and 8 thrusts back, I can get to a place exactly 1 x 13mm length from the start.

That's 15 thrusts.

Is this the minimum number of steps to get back to this minimum distance from the start?

We can check with a little python program:

```python
import math
min = math.inf
for n in range (16):
    for m in range (16):
        if 31*n - 27*m == 1:
            if n + m < min:
                min = n + m
                a = n
                b = m
print(f'n={a}, m={b}')    
```

So, yes, 15 is the minimun.

What about it we move back further, to -1 length (-13mm) from the start? Can I do that in less than 15 thrusts?

```python
min = math.inf
for n in range (15):
    for m in range (15):
        if 31*n - 27*m == -1:
            if n + m < min:
                min = n + m
                a = n
                b = m
if min < math.inf:
    print(f'n={a}, m={b}') 
else:
    print("No solution found less than 15")
```

No we can't.

13mm is the closest we can get to the landing mark without actually hovering directly over the start line, and 15 thrusts in the minimum number to get there.
